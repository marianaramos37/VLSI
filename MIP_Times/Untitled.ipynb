```python
stat = "time="

timesr = []
timess = []

for x in range(1,39):
    f = open('ins{}MIP.txt'.format(x),'r')
    txt = f.read()
    start = txt.find(stat)
    newtxt = txt[start:]
    end = newtxt.find("\n")
    timesr.append(float(newtxt[len(stat):end]))
    print(str(x)+",",newtxt[len(stat):end])
    
    f = open('ins{}MIP_Rotation.txt'.format(x),'r')
    txt = f.read()
    start = txt.find(stat)
    newtxt = txt[start:]
    end = newtxt.find("\n")
    timess.append(float(newtxt[len(stat):end]))
    print(str(x)+",",newtxt[len(stat):end])
```

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# set plot style: grey grid in the background:
sns.set(style="darkgrid")

# load dataset
tips = sns.load_dataset("tips")

# Set the figure size



plt.figure(figsize=(14, 10))

# plot a bar chart
ax = sns.barplot(
    y=timesr, 
    x=list(range(1,39)),
    estimator=sum, 
    ci=None, 
    color='#69b3a2');

ax.set_yscale("log")
ax.axhline(300, color = 'red', ls = "--")

ax.set(xlabel='Instances', ylabel='Time in Seconds', title="Time per Instance")
plt.show()
```
